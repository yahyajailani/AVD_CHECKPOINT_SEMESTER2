### **Outliers Values**

Pada tahap ini, kita akan melihat nilai ekstrem (outliers) pada kolom yang bertipe data numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Akan tetapi memang tidak semua kolom perlu kita tangani, kita harus tetap melihat dan memahami tujuan dari setiap kolom yang ada di dataset.

Dan terlihat kolom

Previous Purchases

Dengan presentase sekitar 4%

Purchase Amount (USD)

Dengan presentase sekitar 0% (sangat kecil / mendekati 0%)

Age

Dengan presentase 0%

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

#Dataframe dari list hasil

results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

#Tampilkan dataframe

display(results_df)
```

# **Eksplorasi Data EDA**

## **Comparison(Perbandingan)**

Aktivitas: Membandingkan total nominal pembelian (Purchase Amount (USD)) di setiap kategori produk (Category).

Tujuan: Mengidentifikasi kategori produk mana yang memberikan kontribusi pendapatan terbesar dan membandingkan performa antar kategori secara langsung.

Visualisasi: Bar Chart (Grafik Batang).

```python
penjualan_category = df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))

sns.barplot(x=penjualan_category.index, y=penjualan_category.values, palette='viridis', hue=penjualan_category.index, legend=False)
plt.title('Total Purchase Amount Berdasarkan Category')
plt.xlabel('Category')
plt.ylabel('Total Purchase Amount (USD)')
plt.xticks(rotation=45)
plt.show()
```

Secara keseluruhan, terdapat perbedaan kontribusi pendapatan yang cukup jelas antar kategori. Strategi bisnis dapat difokuskan pada mempertahankan dominasi Clothing sekaligus meningkatkan penjualan pada kategori dengan kontribusi lebih rendah untuk menyeimbangkan performa penjualan.

## **Composition(Komposisi)**

Aktivitas: Melihat proporsi atau persentase total pembelian (Purchase Amount (USD)) berdasarkan kategori produk (Category).

Tujuan: Mengetahui kategori produk mana yang paling mendominasi total pembelian pelanggan secara keseluruhan.

Visualisasi: Pie Chart (Grafik Lingkaran).

```python
purchase_by_category = df.groupby('Category')['Purchase Amount (USD)'].sum().sort_values(ascending=False)
purchase_by_category_top3 = purchase_by_category.head(3)

plt.figure(figsize=(10, 8))
purchase_by_category_top3.plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Proporsi 3 Kategori Teratas Berdasarkan Total Purchase Amount')
plt.ylabel('')
plt.axis('equal')
plt.show()
```

Dominasi Clothing dalam komposisi total pembelian mengindikasikan bahwa strategi bisnis sebaiknya tetap mempertahankan ketersediaan dan variasi produk pada kategori ini, sekaligus meningkatkan promosi pada kategori lain agar distribusi kontribusi pendapatan menjadi lebih seimbang.

Outlier merupakan nilai dalam suatu dataset yang memiliki perbedaan sangat jauh dibandingkan sebagian besar data lainnya. Nilai ini biasanya tampak terlalu tinggi atau terlalu rendah dari pola umum yang terbentuk, sehingga dapat memengaruhi hasil analisis, khususnya pada perhitungan statistik seperti rata-rata dan standar deviasi.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

Terlihat bahwa pada kolom Customer ID, Age, Purchase Amount (USD), Review Rating, dan Previous Purchases tidak ditemukan outliers (0.0). Hal ini menunjukkan bahwa seluruh nilai pada kolom tersebut masih berada dalam batas yang normal dan tidak ada data yang terlalu ekstrem dibandingkan data lainnya. # **Construct Data**

Tahap ini dilakukan dengan menambahkan kolom baru yang masih berhubungan dengan data sebelumnya, sehingga informasi yang dihasilkan tetap saling berkaitan dan lebih mudah digunakan dalam proses analisis.

Membuat Kolom Baru “Customer Segment” hal ini perlu ditambahkan untuk mengelompokkan pelanggan berdasarkan karakteristik tertentu seperti jumlah pembelian atau nilai transaksi. Dengan adanya segmentasi ini, analisis perilaku belanja menjadi lebih terarah dan visualisasi lebih mudah dipahami saat melihat pola serta tren konsumen.


```python
def customer_segment(previous_purchases):
    if previous_purchases == 0:
        return 'New Customer'
    elif previous_purchases <= 5:
        return 'Regular Customer'
    else:
        return 'Loyal Customer'

df['Customer Segment'] = df['Previous Purchases'].apply(customer_segment)
```
